### This creates a single realistic transaction dict using:
	â€¢	names
	â€¢	cities
	â€¢	merchants
	â€¢	addresses

```python
import pandas as pd

names = pd.read_csv("data/names_extended.csv")
cities = pd.read_csv("data/cities_extended.csv")
merchants = pd.read_csv("data/merchants_extended.csv")
addresses = pd.read_csv("data/addresses_extended.csv")

# Clean column names for ALL dataframes
names.columns = names.columns.str.strip()
cities.columns = cities.columns.str.strip()
addresses.columns = addresses.columns.str.strip()
merchants.columns = merchants.columns.str.strip()
```

```python
import random
import uuid
import time

def generate_transaction():

    user = names.sample(1).iloc[0]
    city = cities.sample(1).iloc[0]["city"]
    addr = addresses.sample(1).iloc[0]["address"]
    merchant = merchants.sample(1).iloc[0]

    return {
        "transaction_id": str(uuid.uuid4()),
        "user_name": user["name"],
        "email": user["email"],
        "phone": user["phone"],

        "merchant": merchant["merchant"],
        "category": merchant["category"],

        "amount": round(random.uniform(99, 9999), 2),
        "timestamp": int(time.time() * 1000),

        "city": city,
        "address": addr
    }

generate_transaction()
```
